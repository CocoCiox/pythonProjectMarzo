```python
import pandas as pd
import matplotlib.pyplot as plt

unicorn = pd.read_csv("unicorn.csv")
unicorn.head()
```

```python
unicorn_states = unicorn[unicorn["Country"] == "United States"]
unicorn_states.head()
```

```python
unicorn_states
```

```python
unicorn_states_data = unicorn_states[unicorn_states["Category"] == "Data management & analytics"]
unicorn_states_data.head()
```

```python
tab = unicorn_states_data.iloc[:, 0:2]
tab_sorted = tab.sort_values('Valuation ($B) ')
tab_sorted
```

```python
tab_sorted.plot(x = 'Company', y = 'Valuation ($B) ', title = 'Unicorn Agencies in the States', rot = 90, kind = 'scatter')
plt.ylim([0, 40])
plt.yticks(range(0, 41, 5))
plt.show()
```
